```python
import requests 
from bs4 import BeautifulSoup 

def get_list_of_links_from_url(url):
    try:
        r = requests.get(url) 
        html_doc = r.text 

         # Use the Beautiful Soup library to parse the HTML document and create a tree structure from it.  
        soup = BeautifulSoup(html_doc, 'html.parser')

         # Create an empty list called links which will store all of our scraped URLs:   
        links = []

         # Use methods of the soup object to find all instances of the <a> tag in order to scrape all URLs on the page:    
        for link in soup.find_all('a'):     
            if 'href' in str(link):
                child = link['href']     # Get the href attribute of each <a> tag and store it in a variable    
                if 'https' in str(child):
                    links.append(child)      # Append each URL found onto our list of links  

        return links
    except:
        return []
                
# Print out all URLs found on this page:
origin = 'https://www.macrotrends.net/stocks/charts/AAPL/apple/revenue'
links = get_list_of_links_from_url(origin)

queue = [origin]
mapping = {}

while len(queue) != 0:
    url = queue.pop(0)
    if url not in mapping:
        links = get_list_of_links_from_url(url)
        mapping[url] = links
        
        for link in links:
            if link not in mapping:
                queue.append(link)
```
